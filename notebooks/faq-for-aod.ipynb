**What is the difference between supervised and unsupervised machine learning?**  
-   *Supervised:* learning labels are provided which is used to train the model to fit to the provided labels. It is able to from an unknown piece of data predict what class it should belong to if there are classes, if it is continuous it will predict a value (e.g. regression)
-   *Unsupervised:* learning no labels are provided, the data is instead clustered together in different classes. It is able to find structure and relationships between items.

**What is over-fitting, how do I know if I have it, and how do I fix it?**  
Too many parameters that has been trained for too long, without using the correct techniques, the end result it works great for the specific training set but not for the general patterns. You would see his higher accuracy for the training / test set but lower against the validation set.  
There are various techniques for fixing overfitting:
-   *Use dropout:* This can added at each layer of the model it will mean that a percentage of the activations will be removed at random so that the model can not overfit based on the data even if it is trained for a long time.
-   *Data augmentation*: Change the data in subtle ways at random to reflect real world scenario's, e.g. if doing image detection, slightly rotate, crop, pan etc the image when doing the training differently at each epoch.
-   *Do not over train the model:* Too many epochs can cause the model to overfit as it will be training too much on the training and test sets.

## Deep Learning


**How do you compare Tensor Flow and MXNet?**  
MxNet has near linear horizontal scaling, and the
models can be distributed on commodity or low powered hardware for predictions. MxNet has significant performance at scale compared to other libraries, it is incubated by Apache by multiple backers from industry and academia.

**How do you know which neural network to build?**  
It depends on the end goal you are trying to achieve and what the data is some examples are:
-   CNN are good for images they model well a representation of spacial relationships, its all about how items related to each other from the point of view of position.
-   RNN are good for NLP type tasks where memory and state key to keeping track of short term and long term values.

**What is the meaning of the following deep learning terms: CNN, RNN, LSTM, Embedding, Auto-Encoders, Dropout?**

-   *RNN:* Recurrent Neural Network: Have the concept of sequence and a memory that can be used to remember state and use that for learning.
-   *LSTM:* Long Short Term Memory: A type of RRN that specializes in remembering values for either a long or short period of time.
-   *Embedding:* Converting entities (often words) to an key value lookup, e.g. take a bag of words and assign each unique word an ID and use that array for training the model. Alternatively it could be an entity e.g. a movie ID then it would return the movie at X index.


**Why are GPU's important for Deep Learning neural networks?**  
Deep learning has a huge amount of parameters and requires a long time for a model to train on a standard CPU, even similar models. Now models can have more layers and be more complex as the GPU offers multiple order of magnitude more computation that is needed to allow a model to be trained in an acceptable amount of time.

**How does MXNet help with implementing Deep Learning neural networks?**  
MxNet key features are:
-   Highly scalable: parallelized compute and I/O operations
-   Wide range of programing languages integrations
-   Portable runs on many different types of platform, GPU, CPU, mobile or IoT devices.

**Explain (at a high level) how Deep Learning neural networks are trained and how they learn.**  
Deep learning is an algorithm modeled on how the human brain works.  The more data given and more compute it has the better it gets, potentially it does not have any limitations in the type of tasks it can solve.  The model is made of layers each having many activations (which can be thought of as a neurons similar to human brains) these have different values that are trained from adding more and more data.  The training process is heavy on compute as it is looking to fit the activation values based on the training data then this is tested and validated with test and validation data.

## Rekognition
**What are the main usages of Rekognition?**  
Rekognition has several main use cases for analyzing pictures:
-   *Scene detection:* provide labels and scores describing the scene or objects. The confidence score can be used to be more sensitive to false positives.
-   *Face analysis:* Detect up to 15 faces in an image and extract relevant features from it e.g. gender, smile, age, glasses, eyes open, mouth open, quality etc
-   *Face comparism:* Search against a set of faces to see how similar a particular face is against the dataset, it will give you a likely hood of match and bounding box of the detected face.

**What kind of labels do you have in Rekognition, and what should I do if you don't recognize my labels?**  
Rekognition has thousands of labels for most common categories,  such as people, food, sports, nature, transport, electrical, animals and many more.  More labels are added largely based on customer demand, if you find a label that is not detected please contact our support team.

**Can you do OCR? Can you do content filtering?**
OCR is not currently supported by Rekognition, content moderation is now available.

**Do you keep my images when I'm sending them to labeling, or face searches?**  
No, not unless you give us permission to store them.   When doing face searches rekognition stores only metadata that allows it to do the search not the image.

**How can I build a security system that will be used in my company gates?**  
You can use facial comparison to see if a photo matches one that is on record, this should however be used as part of a multi factor authentication rather than the single authentication method.

## Polly
**What are the main usages of Polly?**  
Polly creates an audio file from text, the audio is life like speech. It can be used anywhere that speech could be useful, reading out news or books, as part of a chat bot, hands free operation.  Polly generates audio in real time creating use cases for interactive applications.  Audio files can be saved or cache saving the need to re-generate the same audio multiple times or the files can be stored on mobile / IoT devices to alert about state or talk to the user instead of beeping or flashing.  User manuals can be embedded in speech.  It currently has 47 voices in 23 languages, users of different languages can have the same experience.

**Can I control the pitch, speed or volume of Polly?**  
With Speak Synthesis Markup Language (SSML) you can control, the pitch, speed, volume, language, paragraphs, plus other attributes.

**How can I add voice to my mobile application? What if it doesn't have connectivity in times?**  
The mobile device can talk directly to Polly to get real time responses from the service for dynamic in app content,.. Common audio responses can be stored on the device, giving a voice experience when offline.   SDK's for android and IOS are available to simplify interaction with the service from mobile, other SDK's are available for other types of integrations.

**What can I do if Polly doesn't know the names of my products (medicies, for example)?**  
You create customer lexicons that allow control of the pronunciation of specific terms. Lexicons are often used for words stylized with numbers to read them as a human would or for acronym to be spelled out.

## Lex
**Why do I need a chatbot?**  
Chats bots are the interface of the future, in many use cases they are the preferred method for human interaction, they are flexible and allow interacting with they user while they are multitasking or in environments where it would not normally be possible.  The input to a chat box in virtually infinitely flexible allowing the user to express their need in the way they want.

**What is the difference between Alexa and Lex, and when should I use which?**  
Alexa allows you to create skills with then are install able by users and usable on any Alexa enabled device (including non Amazon devices with Amazon Voice Service installed). Lex give you the ability  to build your own chat bots in a similar way outside of the Alexa ecosystem. You create intents and slots in the same way, however you can have it privately available or in your app or interact with text or voice, you have more control.

**Can I use Lex for speech recognition?**  
Lex performs speech recognition as part of its Natural Language Understanding (NLU) however it is not broken out as a separate service.

**How can I build a system that will help in my customer support (for example, AWS technical support)?**  
Different intents can be defined, e.g. limit increase, service request, how to, break fix, and slot types defined for each of them, utterances would be created so that when someone is talking to the bot it would know what intent is most relevant and what slots should be filled.  The bot could respond where the resolution is well defined e.g. limit increases, or where it is more complex it can collect more information and pass to a human representative.

## Amazon Machine Learning (AML)
**When should I use Amazon ML?**  
The ML service is a great starting point for machine learning, allowing you to train a model and then deploy the model that can be predicted at scale without having to manage this process. It is well suited for the long tail of machine learning problems that your organization would have.

**Which of the classic machine tasks are addressed by AML?**  
AML deals with the long tail of machine learning tasks, specifically:
-   *Regression:* predict a numeric value, e.g. days to keep an item
-   *Binary classification:* Is it a YES or NO
-   *Multi-class classification:* A,B,C,D or which team should it go to

**What do I need to do to improve the performance of my ML model?**  
Visualization tools are provided to evaluate the model's behavior, prediction results can be visualized meaning you can fine tune the model performance based on thresholds and preference of false positives vs false negatives or overall performance.

## Machine Learning on Amazon EMR
**What is the key application to use for ML on EMR?**  
Apache Spark ML is the key application for machine learning on EMR, it specializes in machine learning at scale.  It is efficient use of cache gives it speed improvements of up to 100x compared to Hadoop (depending on the workload) machine existing algorithms can be included via MLLib.

**When would you use EMR-based ML (Spark) instead of AML?**  
AML is best for long tail problems or where you don't want to put data science resource or you quickly need to get something in to production at scale. Spark ML on EMR best when you want to have full control over the model that is being created yet in a managed cluster.

**What additional work would you need to do to properly leverage Spark ML that AML already provides for you?**  
Making the model available for prediction at production scale to meet demand is provided automatically in AML for Spark ML much of this will need to be implemented.

**What is a Spark ML Pipeline?**  
Pipelines in Spark ML make it easier to combine multiple algorithms into a single workflow. The API is unified simplifying this process.

**What do the Transformer, Estimator, and Model components of a Spark ML Pipeline do?**  
-   *Dataframe:* Hold data of various types.
-   *Transformer:* Algorithm that changes data in a dataframe from one dataframe to another.
-   *Estimator:* Algorithm that with learns from the data in a dataframe outputting a model.
